```python
from graphviz import Digraph

def trace(root):
  # builds a set of all nodes and edges in a graph
    nodes, edges = set(), set()
    
    
    def build(v):
        if v not in nodes:
            nodes.add(v)
            for child in v.children:
                edges.add((child, v))
                build(child)
                
    build(root)
    return nodes, edges

def drawgraph(root):
    dot = Digraph(format='png', graph_attr={'rankdir': 'LR'}) # LR = left to right
    nodes, edges = trace(root)
    for n in nodes:
        uid = str(id(n))
        dot.node(
            name=uid,
            label = "{%s | data= %.4f | grad= %.4f}" % (n.label, n.data, n.grad),
            shape='record'
        )
        
        if n.op:
            dot.node(
                name=uid + n.op,
                label=n.op
            )
            dot.edge(uid + n.op, uid)
    
    for n1, n2 in edges:
        dot.edge(
            str(id(n1)),
            str(id(n2)) + n2.op
        )
    return dot
```

```python
class Neuron1:
    
    def __init__(self, input_length: int):
        self.input_length = input_length
        self.w = [random.random() for i in  range(self.input_length)]
    
    def activation(self, inp):
        return math.tanh(inp)
        
    def __call__(self, xinput: list, b=0.0):
        out = 0
        for x, w in zip(xinput, self.w):
            out += x*w
        out += b
        
        return self.activation(out)

    
    
random.seed(22)

h = 0.0001
x1 = 0.1 + h
x2 = -0.3
x3 = 0.2


x = [x1, x2, x3]
b = 1
n1 = Neuron1(len(x))

y = n1(x, b)
y
```

```python
class Neuron:
    def __init__(self, xlen: int):
        self.w = [Value(random.uniform(-1, 1), "w{}".format(i)) for i in range(xlen)]
        self.b = Value(random.uniform(-1, 1), "b")

    def __call__(self, inputs: list):
        
        oval = 0.0
        for x, w in zip(inputs, self.w):
            oval += x * w
        
        oval += self.b
        
        oval.backprop()
        
        
        
x = [1.0, 3.0, 2.0]
n = Neuron(len(x))
n(x)

```
